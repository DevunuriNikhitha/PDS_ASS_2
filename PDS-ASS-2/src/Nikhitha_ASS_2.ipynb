```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('train.csv')
data
```

```python
data.info()
```

In the 14 attributes there are 5 attributes with the null values

```python
missing = data.isnull().sum()
missing
```

```python
#mileage median
```

Mileage is the attribute with the numerical type,so used median to fill the missing values ,which is effective instead of droping because there is small number of missing values i.e;2.
Firstly removed the units for the values in the mileage attribute then calculated median and filled the value in the missing place

```python
data['Mileage'] = data['Mileage'].str.replace(r'kmpl', '', regex=True)
data['Mileage'] = data['Mileage'].str.replace(r'km/kg', '', regex=True)
```

```python
mileage_median = data['Mileage'].median()
mileage_median
```

```python
data['Mileage'].fillna(mileage_median, inplace=True)
data
```

```python
missing = data.isnull().sum()
missing
```

```python
#seats mode

```

seats is the attribute with the numerical type,so used mode to fill the missing values ,which is effective instead of mean and median because mode gives the perfect number where as mean, median gives float which is incompatable value for the car seats.

```python
seats_mode = data['Seats'].mode()[0]
seats_mode
```

```python
data['Seats'].fillna(seats_mode, inplace=True)
data
```

```python
missing = data.isnull().sum()
missing
```

```python
#power median
```

power is the attribute with the numerical type,so used median to fill the missing values ,which is effective instead of droping because there is small number of missing values i.e;36.firstly removed the units for the values in the power attribute then calculated median and filled the value in the missing place

```python
data['Power'] = data['Power'].str.replace(r'bhp', '', regex=True)
```

```python
power_median = data['Power'].median()
power_median
```

```python
data['Power'].fillna(power_median, inplace=True)
data
```

```python
missing = data.isnull().sum()
missing
```

```python
#engine mean
```

engine is the attribute with the numerical type,so used mean to fill the missing values ,which is effective instead of droping because there is small number of missing values i.e;36.firstly removed the units for the values in the engine attribute then calculated mean value and filled the value in the missing place

```python
data['Engine'] = data['Engine'].str.replace(r'CC', '', regex=True)
```

```python
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')
```

```python
engine_mean = data['Engine'].mean()
engine_mean
```

```python
engine_mean = int(engine_mean)
engine_mean
```

```python
data['Engine'].fillna(engine_mean, inplace=True)
data
```

```python
missing = data.isnull().sum()
missing
```

```python
#new_price dropping
```

New_price is the attribute with the numerical type,Used dropping the rows because there are 5032 null values,which is effective instead of filling the missing values with the mean,median,mode.firstly removed the units for the values in the new_price attribute then dropped the rows with the missing values.

```python
data['New_Price']=data['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
d1=data.dropna()
d1
```

```python
missing = d1.isnull().sum()
missing
```
